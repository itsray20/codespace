## Problem Statement
Write a program to scheme a few activation functions that are used in neural networks

## Importing required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Creating a dataset

```python
# data has 100 points evenly spaced between -1 and 1
data = np.linspace(-20, 20, 100)
```

## Function for plotting activation functions using MatPlotLib

```python
def plot_function(x, y, x_label='', y_label='', name=''):
    plt.figure(figsize=(10, 10))
    plt.title(f'{name} Activation Function')
    plt.plot(x, y)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.axvline(x=0)
    plt.axhline(y=0)
    plt.show()
```

## Linear Activation Function
### $y = x$


```python
# Plots y = x
plot_function(x=data,
              y=data,
              x_label='x',
              y_label='f(x)',
              name='Linear')
```

## ReLU Activation Function (Rectified Linear Unit)
### $y=\begin{cases}0\text{ if } x < 0\\x \text{ if } x \geqslant 0\end{cases}$

```python
# Plots y = 0 for negative values of x and y = x for the rest
plot_function(x=data,
              y=data * (data > 0),
              x_label='x',
              y_label='ReLU(x)',
              name='Relu')
```

## Softmax Activation Function
### $\sigma(x)=\frac{e^{x}}{\displaystyle\sum_{x=1}^ne^{x}}$

```python
plot_function(x=data,
              y=np.exp(data - data.max()) /
              np.sum(np.exp(data - data.max())),
              x_label='x',
              y_label='softmax(x)',
              name='Softmax')
```

## Sigmoid Activation Function
### $sigmoid(x)=\frac{1}{1+e^{-x}}$

```python
plot_function(x=data,
              y=1 / (1 + np.exp(-data)),
              x_label='x',
              y_label='sigmoid(x)',
              name='Sigmoid')
```

## tanh Activation Function
### $\tanh(x)=\frac{e^{x}-e^{-x}}{e^{x}+e^{-x}}$

```python
plot_function(x=data,
              y=(np.exp(data) - np.exp(-data)) /
              (np.exp(data) + np.exp(-data)),
              x_label='x',
              y_label='tanh(x)',
              name='tanh')
```
